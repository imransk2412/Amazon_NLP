```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary datasets
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Initialize stopwords, stemmer, and lemmatizer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    words = word_tokenize(text)  # Tokenization
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    words = [lemmatizer.lemmatize(word) for word in words]  # Apply lemmatization
    return " ".join(words)

# Apply the function to the 'verified_reviews' column
df['cleaned_reviews'] = df['verified_reviews'].apply(clean_text)

# Display the first few cleaned reviews
df[['verified_reviews', 'cleaned_reviews']].head()
```

```python
from collections import Counter
import seaborn as sns
import nltk

nltk.download('punkt')
```

```python
def get_ngrams(text_column, n=2, top_k=20):
    all_words = " ".join(text_column.dropna()).split()
    ngrams = list(nltk.ngrams(all_words, n))
    ngram_freq = Counter(ngrams)
    return ngram_freq.most_common(top_k)
```

```python
bigrams = get_ngrams(df['cleaned_reviews'], n=2)

# Convert to DataFrame
bigram_df = pd.DataFrame(bigrams, columns=['Bigram', 'Count'])

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=bigram_df['Count'], y=[" ".join(x) for x in bigram_df['Bigram']], palette="viridis")
plt.xlabel("Frequency")
plt.ylabel("Bigrams")
plt.title("Top Bigrams in Reviews")
plt.show()
```

```python
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity  # Returns value between -1 (negative) and +1 (positive)

# Apply function
df['sentiment'] = df['cleaned_reviews'].apply(get_sentiment)

# View sentiment distribution
df[['cleaned_reviews', 'sentiment']].head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['sentiment'], bins=30, kde=True, color="blue")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.title("Sentiment Distribution of Reviews")
plt.show()
```
